## Bygg korpus

Korpuset er en definisjon av et korpus, og består av en samling URN-er.


Korpuset defineres med metadata som dewey, emneord, navn , år, etc. Her kan [Webdewey](http://deweysearchno.pansoft.de/webdeweysearch/index.html) være til god hjelp.

Kommandoen er `book_urn()`

```python
korpus = nb.book_urn( lang =  'sme', limit=2000)
korpus[:10]
```

## Konkordans

Kommando for å lage konkordanser fra et sett med URN-er er `urn_concordance()` (OBS! Korpuset bør være under 100 bøker for responsivitet)

```python
nb.urn_concordance(word = 'odda', urns = korpus, before = 10, after = 10)
```

```python
import requests
def iiif_manifest(urn):
    r = requests.get("https://api.nb.no/catalog/v1/iiif/{urn}/manifest".format(urn=urn))
    return r.json()

def pages(urn, scale=800):
    a = iiif_manifest(urn)
    return [page['images'][0]['resource']['@id'].replace("full/full/", "full/0,{s}/".format(s=scale)) for page in a['sequences'][0]['canvases']]

def load_picture(url):
    r = requests.get(url, stream=True)
    r.raw.decode_content=True
    return r.raw

def get_url(urn, page=1, part=200):
    # urn as digit
    urn = "URN:NBN:no-nb_digibok_" + urn + '_{0:04d}'.format(page)
    print(urn)
    url = "https://www.nb.no/services/image/resolver/{urn}/full/0,{part}/native.jpg".format(urn = urn, part=part)
    return url


def page_urn(urn, page=1):
    # urn as digit
    return "URN:NBN:no-nb_digibok_" + urn + '_{0:04d}'.format(page)
```
